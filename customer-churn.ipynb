```python
import pandas as pd
```

```python
df = pd.read_csv("Customer-Churn-Records.csv")
df.head()
```

```python
df.columns
```

```python
df_clean = df.drop(columns=["RowNumber", "CustomerId", "Surname"])

df_clean["Churn Status"] = df_clean["Exited"].map({
    1: "Churned",
    0: "Retained"
})

df_clean["Complaint Status"] = df_clean["Complain"].map({
    1: "Complained",
    0: "No Complaint"
})

df_clean["Active Status"] = df_clean["IsActiveMember"].map({
    1: "Active",
    0: "Inactive"
})

df_clean["Credit Card Status"] = df_clean["HasCrCard"].map({
    1: "Has Credit Card",
    0: "No Credit Card"
})

df_clean["Age Group"] = pd.cut(
    df_clean["Age"],
    bins=[0, 30, 45, 60, 100],
    labels=["Under 30", "30-45", "46-60", "60+"]
)

df_clean["Balance Group"] = pd.cut(
    df_clean["Balance"],
    bins=[-1, 0, 50000, 150000, 300000],
    labels=["No Balance", "Low Balance", "Medium Balance", "High Balance"]
)

df_clean["Credit Score Group"] = pd.cut(
    df_clean["CreditScore"],
    bins=[0, 580, 670, 740, 850],
    labels=["Poor", "Fair", "Good", "Excellent"]
)

df_clean.head()
```

```python
df_clean.to_csv("cleaned_customer_churn.csv", index=False)
```
